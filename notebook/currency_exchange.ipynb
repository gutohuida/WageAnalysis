```python
CURRENCY = "BRL"
```

```python
import requests
usd_exchange = requests.get(f"https://open.er-api.com/v6/latest/{CURRENCY}")
```

```python
print(usd_exchange.content.decode('utf-8'))
```

```python
import ast
import json
print(json.dumps(ast.literal_eval(usd_exchange.content.decode('utf-8')), indent=4))
```

```python
result_dict = ast.literal_eval(usd_exchange.content.decode('utf-8'))
```

```python
import pandas as pd
from datetime import datetime, timezone
currency_exchange = pd.DataFrame()

for key in result_dict["rates"]:
    line = {
        "currency_from": key,
        "currency_to": CURRENCY,
        "exchange_rate": result_dict["rates"][key],
        "insert_date": datetime.now().astimezone(timezone.utc)
    }
    currency_exchange = pd.concat([currency_exchange, pd.DataFrame(line, index=[0])], ignore_index=True)
    #print(line)
currency_exchange.head(10)    
```

```python
from sqlalchemy import create_engine
postgresql_url = 'postgresql+psycopg2://postgres:postgres@localhost:5432/wageanalysis'
engine = create_engine(postgresql_url)
```

```python
currency_exchange.to_sql('currency_exchange', engine, schema='raw', if_exists='append', index=False)
```
