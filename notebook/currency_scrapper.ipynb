```python
import pycountry
import pandas as pd
from sqlalchemy import create_engine
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
postgresql_url = 'postgresql+psycopg2://postgres:postgres@localhost:5432/wageanalysis'
engine = create_engine(postgresql_url)
```

```python
driver = webdriver.Chrome()
driver.get('https://www.iban.com/currency-codes')
```

```python
page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')
driver.close()
```

```python
table = soup.find('table', class_='table table-bordered downloads tablesorter').find('tbody')
print(table)
```

```python
df_countrys = pd.DataFrame()
dict_countrys = {}
```

```python
lines = table.find_all('tr')
for line in lines:
    all_data = line.find_all('td')
    dict_countrys = {
        "name": all_data[0].text.lower(),
        "currency_name": all_data[1].text,
        "currency_code": all_data[2].text
    }
    df_countrys = pd.concat([df_countrys,pd.DataFrame(dict_countrys, index=[0])], ignore_index=True)    
    for data in all_data:
        print(data.text)
    print('*******')    
```

```python
df_countrys.to_sql('countrys', engine, schema='raw', if_exists='replace', index=False)
```
